### Configuration File Handling in Python

##### What is a Configuration File?
- A configuration file stores settings or parameters of a program.
- It helps to avoid hardcoding values in code.
- Example: database URL, API keys, file paths, etc.

#### Why use Configuration Files?
- Easy to change settings without changing code
- Keeps code clean and reusable
- Helps in security (store secrets outside code)
- Useful for different environments (dev, test, prod)

#### Common Types of Config Files in Python
- .ini → simple key-value format
- .json → structured data (most common)
- .yaml → human readable (used in DevOps)
- .env → store environment variables

##### file modes

- "r" - Read - Default value. Opens a file for reading, it returns an error if the file does not exist
- "w" - Write - Opens a file for writing, creates the file if it does not exist

#### Best Practices
- Do not store passwords in code
- Use .env for secret keys
- Use JSON/YAML for structured data
- Keep config files separate from code
- Use .gitignore to hide sensitive files

#### Cheat Sheet

| File Type | Import | Read Syntax | Write Syntax |
|----------|--------|------------|--------------|
| JSON | `import json` | `data = json.load(file)` | `json.dump(data, file, indent=4)` |
| INI | `import configparser` | `config.read("file.ini")`<br>`config["SECTION"]["key"]` | `config["SECTION"] = {"key": "value"}`<br>`config.write(file)` |
| .env | `from dotenv import load_dotenv`<br>`import os` | `load_dotenv()`<br>`os.getenv("KEY")` | Not written using Python usually (manual file edit) |
| YAML | `import yaml` | `data = yaml.safe_load(file)` | `yaml.dump(data, file)` |

16. Write a Python program to calculate the total number of bricks required for a 4BHK flat, where each wall has length 15 ft and height 10 ft, and each brick has length 1 ft and height 0.5 ft. Consider shared walls between rooms. Also calculate the total cost using brick price from a config file.

```python
import configparser

config = configparser.ConfigParser()
config.read(r"C:\Users\anike\OneDrive\Desktop\Python\Python Fundamental\data\config3.ini")

def total_num_of_bricks_for_4BHK(wall_length, wall_height, brick_length, brick_height):
    area_of_single_wall = wall_length * wall_height
    total_walls = (4 * 4) - 4    #(rooms × 4) − shared walls
    total_wall_area = area_of_single_wall * total_walls
    brick_area = brick_length * brick_height
    total_bricks = total_wall_area / brick_area
    return total_bricks

def total_cost_of_bricks(total_bricks_count):
    brick_cost = float(config["raw_materials"]["brick_cost"])
    return total_bricks_count * brick_cost

total_bricks_count = total_num_of_bricks_for_4BHK(15, 10, 1, 0.5)
total_cost = total_cost_of_bricks(total_bricks_count)
# We pass total_bricks_count as an argument to the function, and the returned result is stored in the total_cost variable.

print(f"Total num of bricks for 4 BHK are {round(total_bricks_count)}")
print(f"Total cost of bricks for 4 BHK is {round(total_cost)}")
```

17. Given an INI file containing book prices and a dictionary with details of books purchased by each student, calculate the total cost of books for each student. Apply a 10% discount only if a student buys 2 or more books.

```python
import configparser

config = configparser.ConfigParser()
config.read(r"C:\Users\anike\OneDrive\Desktop\Python\Python Fundamental\data\config4.ini")

student_details = {
    1: ["math", "history"],
    2: ["biology", "chemistry", "history"],
    3: ["science"],
    4: ["science", "math", "history", "physics", "biology", "chemistry"]
}

def calculate_cost(student_details):
    results = {}

    for student, books in student_details.items():
        total_cost = 0

        for book in books:
            total_cost += float(config["books"][book])

        if len(books) >= 2:
            total_cost *= 0.9

        results[student] = total_cost

    return results


final_cost = calculate_cost(student_details)

for student, cost in final_cost.items():
    print(f"The total cost of books for student {student} is {round(cost)}.")
```
