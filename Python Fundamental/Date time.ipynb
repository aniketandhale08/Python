### Python datetime

Python has got datetime module to handle date and time.

```python
import datetime
print(dir(datetime))
```

#### Getting datetime Information

```python
from datetime import datetime
now = datetime.now()
print(now)                   
```

```python
day = now.day        
month = now.month           
year = now.year       
hour = now.hour               
minute = now.minute            
second = now.second
timestamp = now.timestamp()
print(day, month, year, hour, minute)

print('timestamp', timestamp)
print(f'{day}/{month}/{year}, {hour}:{minute}')
```

Timestamp or Unix timestamp is the number of seconds elapsed from 1st of January 1970 UTC.

#### Formatting Date Output Using strftime

**IMP**
- strftime → datetime ➝ string
- strptime → string ➝ datetime

```python
from datetime import datetime
new_year = datetime(2020, 1, 1)
print(new_year) 
```

```python
day = new_year.day
month = new_year.month
year = new_year.year
hour = new_year.hour
minute = new_year.minute
second = new_year.second
print(day, month, year, hour, minute) #1 1 2020 0 0
print(f'{day}/{month}/{year}, {hour}:{minute}')  # 1/1/2020, 0:0
```

```python
from datetime import datetime
now = datetime.now()

t = now.strftime("%H:%M:%S")
print("time:", t)          

time_one = now.strftime("%m/%d/%Y, %H:%M:%S")
print("time one:", time_one)       

time_two = now.strftime("%d/%m/%Y, %H:%M:%S")
print("time two:", time_two)      
```

```python
from datetime import datetime
import time

while True:
    now1 = datetime.now()
    current_time = now1.strftime("%d-%m-%y, %H-%M-%S")
    print(current_time)
    time.sleep(2.5)
```

#### String to Time Using strptime

```python
from datetime import datetime
date_string = "5 June, 2019"
print("date_string =", date_string)
```

```python
date_object = datetime.strptime(date_string, "%d %B, %Y")
print("date_object =", date_object) 
```

#### Using date from datetime

```python
from datetime import date
d = date(2020, 1, 1)
print(d)       
print('Current date:', d.today())  
```

```python
today = date.today()
print("Current year:", today.year)   # 2019
print("Current month:", today.month) # 12
print("Current day:", today.day)     # 5
```

#### Time Objects to Represent Time

```python
from datetime import time
a = time()
print("a =", a)

b = time(10, 30, 50)
print("b =", b)    

c = time(hour=10, minute=30, second=50)
print("c =", c)   

d = time(10, 30, 50, 200555)
print("d =", d)   
```

#### Difference Between Two Points in Time Using

```python
from datetime import date, datetime
today = date(year=2019, month=12, day=5)
new_year = date(year=2020, month=1, day=1)
time_left_for_newyear = new_year - today
print('Time left for new year: ', time_left_for_newyear) 
```

```python
t1 = datetime(year = 2019, month = 12, day = 5, hour = 0, minute = 59, second = 0)
t2 = datetime(year = 2020, month = 1, day = 1, hour = 0, minute = 0, second = 0)
diff = t2 - t1
print('Time left for new year:', diff)
```

```python
from datetime import date, datetime

new_year1 = datetime(2027, 1, 1, 0, 0, 0)
today_date = datetime.today()

differ = new_year1 - today_date
print(differ)
```

#### Difference Between Two Points in Time Using timedelta

```python
from datetime import timedelta
t1 = timedelta(weeks=12, days=10, hours=4, seconds=20)
t2 = timedelta(days=7, hours=5, minutes=3, seconds=30)
t3 = t1 - t2
print("t3 =", t3)
```

1. Get the current day, month, year, hour, minute and timestamp from datetime module

```python
from datetime import datetime

now = datetime.now()

day = now.day
month = now.month
year = now.year

hour = now.hour
minute = now.minute
second = now.second

# Timestamp = total seconds passed since 1 Jan 1970 (UTC)
timestamp = now.timestamp()

print(f"Today's date is {year}-{month}-{day}, {hour}-{minute}-{second}")
print("timestamp:", timestamp)
```

    - Time series analysis
    - To get a timestamp of any activities in an application

**Uses of datetime module**

- **Time series analysis** → used in data analysis to track data over time like stock prices, sales, user activity trends  
- **Logging & timestamps** → store exact time of events in applications (errors, login time, API calls) for debugging and tracking  
- **Scheduling & automation** → run tasks at a fixed time like sending emails, backups, cron jobs  
- **Date calculations** → calculate difference between dates like age, project deadlines, subscription expiry  
- **Formatting date/time** → convert date into readable formats for users (like `14/04/2026, 15:30`)  
- **Parsing date strings** → convert user input string into datetime object for processing  
- **Database operations** → store and query time-based records (created_at, updated_at)  
- **Web applications** → show post time, comments time, last seen, session expiry  
- **Timezone handling** → manage different time zones in global applications  
- **Performance tracking** → measure execution time of code or processes
