### Exception Handling

Python uses try and except to handle errors gracefully. A graceful exit (or graceful handling) of errors is a simple programming idiom - a program detects a serious error condition and "exits gracefully", in a controlled manner as a result. Often the program prints a descriptive error message to a terminal or log as part of the graceful exit, this makes our application more robust. The cause of an exception is often external to the program itself. An example of exceptions could be an incorrect input, wrong file name, unable to find a file, a malfunctioning IO device. Graceful handling of errors prevents our applications from crashing.

We have covered the different Python error types in the previous section. If we use try and except in our program, then it will not raise errors in those blocks.




```python
# syntax
try:
    code in this block if things go well
except:
    code in this block run if things go wrong

````python
# Example

try:
    print(10 + '5')
except:
    print('Something went wrong')
````

In the example above the second operand is a string. We could change it to float or int to add it with the number to make it work. But without any changes, the second block, except, will be executed.

````python
try:
    name = input('Enter your name:')
    year_born = input('Year you were born:')
    age = 2019 - year_born
    print(f'You are {name}. And your age is {age}.')
except:
    print('Something went wrong')
````

In the above example, the exception block will run and we do not know exactly the problem. To analyze the problem, we can use the different error types with except.

In the following example, it will handle the error and will also tell us the kind of error raised.

````python
try:
    name = input('Enter your name:')
    year_born = input('Year you were born:')
    age = 2019 - year_born
    print(f'You are {name}. And your age is {age}.')
except TypeError:
    print('Type error occured')
except ValueError:
    print('Value error occured')
except ZeroDivisionError:
    print('zero division error occured')
````

In the code above the output is going to be TypeError. Now, let's add an additional block:

````python
try:
    name = input('Enter your name:')
    year_born = input('Year you born:')
    age = 2026 - int(year_born)
    print(f'You are {name}. And your age is {age}.')
except TypeError:
    print('Type error occur')
except ValueError:
    print('Value error occur')
except ZeroDivisionError:
    print('zero division error occur')
else:
    print('I usually run with the try block')
finally:
    print('I always run.')
````

#### Packing and Unpacking Arguments in Python

We use two operators:

##### Unpacking

Unpacking Lists

````python
def sum_of_five_nums(a, b, c, d, e):
    return a + b + c + d + e

lst = [1, 2, 3, 4, 5]
print(sum_of_five_nums(lst)) # TypeError: sum_of_five_nums() missing 4 required positional arguments: 'b', 'c', 'd', and 'e'
````

When we run the this code, it raises an error, because this function takes numbers (not a list) as arguments. Let us unpack/destructure the list.

````python
def sum_of_five_nums(a, b, c, d, e):
    return a + b + c + d + e

lst = [1, 2, 3, 4, 5]
print(sum_of_five_nums(*lst))  # 15
````

We can also use unpacking in the range built-in function that expects a start and an end.

````python
numbers = range(2, 7)  # normal call with separate arguments
print(list(numbers)) # [2, 3, 4, 5, 6]

args = [2, 7]
numbers = range(*args)  # call with arguments unpacked from a list
print(numbers)      # [2, 3, 4, 5,6]
````

A list or a tuple can also be unpacked like this:

````python
countries = ['Finland', 'Sweden', 'Norway', 'Denmark', 'Iceland']
fin, sw, nor, *rest = countries
print(fin, sw, nor, rest)   # Finland Sweden Norway ['Denmark', 'Iceland']

numbers = [1, 2, 3, 4, 5, 6, 7]
one, *middle, last = numbers
print(one, middle, last)      #  1 [2, 3, 4, 5, 6] 7
````

Unpacking Dictionaries

````python
def unpacking_person_info(name, country, city, age):
    return f'{name} lives in {country}, {city}. He is {age} year old.'
dct = {'name':'Anu', 'country':'India', 'city':'Pune', 'age':23}
print(unpacking_person_info(**dct))
````

##### Packing
Sometimes we never know how many arguments need to be passed to a python function. We can use the packing method to allow our function to take unlimited number or arbitrary number of arguments.

Packing Lists

````python
def sum_all(*args):
    s = 0
    for i in args:
        s += i
    return s
print(sum_all(1, 2, 3))             # 6
print(sum_all(1, 2, 3, 4, 5, 6, 7)) # 28
````

Packing Dictionaries

````python
def packing_person_info(**kwargs):
    for key in kwargs:
        print(f"{key} = {kwargs[key]}")
    return kwargs

print(packing_person_info(name="Anu",country="India", city="Pune", age=23))
````

````python
fruits = ['banana', 'orange', 'mango', 'lemon', 'lime']                    
vegetables = ['Tomato', 'Potato', 'Cabbage','Onion', 'Carrot']
fruits_and_veges = []
for f, v in zip(fruits, vegetables):
    fruits_and_veges.append({'fruit':f, 'veg':v})

print(fruits_and_veges)
````

### Short Notes

#### Exception Handling – Simple Notes

##### What is Exception Handling?

Exception handling is a way to handle runtime errors without stopping the program suddenly.  
It helps the program run safely even if an error happens.

---

##### What is `try`?

- `try` is used to write risky code.  
- The code inside `try` may cause an error.  
- If error happens, Python looks for `except` block.  

**Meaning:** Try this code.

---

##### What is `except`?

- `except` is used to handle the error.  
- It runs only if error happens inside `try`.  
- It prevents program from crashing.  

**Meaning:** If error happens, handle it here.

---

##### What is `raise` (Most Important)?

- `raise` is used to create an error manually.  
- Python does not create this error automatically.  
- We use `raise` when we check a condition and find something wrong.  

**Meaning:** Throw an error.

---

##### Why Use `raise` If We Have `except`?

This is important:

- `raise` is used to detect and create the error.  
- `except` is used to handle the error.  
- They have different roles.  

Without `raise`:
- We must write print and continue everywhere.  
- Code becomes messy.  

With `raise`:
- Validation logic stays clean.  
- Error handling is done in one place.  
- Code looks professional.  

We use raise when:

- Python does NOT automatically give error
- But logically the input is wrong
---

##### When Does Program Stop?

- If `raise` is not handled → program stops.  
- If `raise` is inside `try` and handled by `except` → program continues.  

---

##### Simple Difference

| Keyword  | Purpose                  |
|-----------|--------------------------|
| `try`     | Write risky code         |
| `except`  | Handle error             |
| `raise`   | Create error manually    |


##### Short Examples

##### 1) Simple Example – `try` & `except`

```python
try:
    num = int("abc")
except:
    print("Invalid number")
```

**What happens?**

- "abc" cannot convert to integer.  
- Error happens.  
- `except` handles it.  
- Program does NOT crash.  

Note: Here we did NOT use `raise`.

---

##### 2) Example – Using `raise` (Manual Error)

```python
age = -5

if age < 0:
    raise ValueError("Age cannot be negative")
```

**What happens?**

- Python normally allows -5.  
- We manually check the condition.  
- We say this is wrong.  
- So we `raise` an error.  
- Program stops because error is not handled.  

---

##### 3) `raise` + `try` + `except` Together

```python
try:
    age = -5

    if age < 0:
        raise ValueError("Age cannot be negative")

    print("Valid age")

except ValueError as e:
    print(e)

print("Program continues")
```

#### Common Python Errors (Simple Notes)

##### 1. ValueError
- **When it happens:** Value type is correct but the value is invalid.  
- **Example:** `int("abc")`  
- **Message you can print:** "Please enter a valid value"

---

##### 2. TypeError
- **When it happens:** Operation between incompatible types.  
- **Example:** `"10" + 5`  
- **Message:** "Invalid data type used"

---

##### 3. IndexError
- **When it happens:** List index is outside the valid range.  
- **Example:** `lst[10]` when list has 3 elements  
- **Message:** "Index is out of range"

---

##### 4. KeyError
- **When it happens:** Accessing a dictionary key that does not exist.  
- **Example:** `dct['salary']` when key not present  
- **Message:** "Key does not exist in dictionary"

---

##### 5. ZeroDivisionError
- **When it happens:** Division by zero.  
- **Example:** `10 / 0`  
- **Message:** "Cannot divide by zero"

---

##### 6. FileNotFoundError
- **When it happens:** Trying to open a file that does not exist.  
- **Example:** `open("abc.txt")`  
- **Message:** "File not found"

---

##### 7. NameError
- **When it happens:** Using a variable that is not defined.  
- **Example:** `print(age)` when `age` not declared  
- **Message:** "Variable is not defined"

---

##### 8. AttributeError
- **When it happens:** Object does not have the attribute or method.  
- **Example:** `"abc".append(5)`  
- **Message:** "Attribute does not exist"

---

##### 9. ImportError
- **When it happens:** Python cannot import a module.  
- **Example:** `import abcxyz`  
- **Message:** "Module cannot be imported"

---

##### 10. IndentationError
- **When it happens:** Incorrect indentation in code.  
- **Example:** Wrong spacing after `if`, `for`, etc.  
- **Message:** "Incorrect indentation"

---

##### 11. StopIteration
- **When it happens:** Iterator has no more items.  
- **Example:** `next()` on finished iterator  
- **Message:** "No more elements in iterator"

2. Write a porgram to open a file.
If file is empty then raise an exception.
If it's not empty then find out all of the unique names in the file.

````python
try:
    with open("data.txt", "r") as file:
        content = file.read().strip()

        if not content:
            raise Exception("File is empty")

        lines = content.split("\n")
        unique_names = set()

        for i in range(1, len(lines)):
            parts = lines[i].split(",")

            try:
                id_val = parts[0].strip()
                name = parts[1].strip()
                age = parts[2].strip()
                gender = parts[3].strip()

                if not id_val.isdigit():
                    raise ValueError

                if not (name.startswith('"') and name.endswith('"')):
                    raise ValueError

                if not age.isdigit():
                    raise ValueError

                if not (gender.startswith('"') and gender.endswith('"')):
                    raise ValueError

                name = name.replace('"', '')
                unique_names.add(name)

            except:
                print(f"Skipping invalid row: {lines[i]}")
                continue

        print("Unique names:")
        for name in unique_names:
            print(name)

except FileNotFoundError:
    print("File not found")
except Exception as e:
    print(e)
````

15. You are reading a CSV file. Some rows have missing values. Write logic to skip rows with errors instead of stopping the program.

````python
file = None

try:
    file = open("data.csv", "r")
    lines = file.readlines()

    header = lines[0].strip().split(",")
    expected_columns = len(header)

    print(header)

    for i in range(1, len(lines)):
        try:
            parts = lines[i].strip().split(",")

            if len(parts) != expected_columns:
                raise ValueError

            print(parts)

        except:
            print("Skipping invalid row:", lines[i].strip())
            continue

except Exception as e:
    print(e)

finally:
    if file:
        file.close()
        print("\nProcess completed!!!")
````
