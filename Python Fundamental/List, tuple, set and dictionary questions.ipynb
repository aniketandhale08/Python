### Python Interview Questions (List, Tuple, Set, Dictionary)

Q1. What is dictionary in Python?

Example: `{ "name": "Aniket", "age": 22 }`

Q2. Is dictionary mutable?

Yes, dictionary is mutable. We can add, update, or delete elements.

Q3. Can we use tuple as a key in dictionary?

Yes, tuple can be used as a key because it is immutable.

Q4. Difference between `list.pop()` and `dictionary.pop()`?

`list.pop()` removes an element using index.
`dictionary.pop()` removes an element using key.

Q5. How to get all the keys available in dictionary?

Use `dict.keys()` method.

Q6. Is list mutable?

Yes, list is mutable and elements can be modified.

Q7. Is list ordered?

Yes, list maintains insertion order.

Q8. What is list index out of range?

It is an error that occurs when accessing an index that does not exist.

Q9. How to get last element of a list?

Use negative indexing: `list[-1]`

Q10. How list comprehension works?

It creates a list in a single line using a loop and condition.

Q11. Is tuple mutable?

No, tuple is immutable.

Q12. Is tuple ordered?

Yes, tuple maintains order.

Q13. How to access element of tuple?

Use index value: `tuple[index]`

Q14. How to add element in tuple?

Directly not possible. Convert tuple to list, add element, then convert back.

Q15. How to reverse item of a tuple?

Use slicing: `tuple[::-1]`

Q16. Is set mutable?

Yes, set is mutable.

Q17. Is set ordered?

No, set is unordered.

Q18. If list is converted into set then will order be maintained?

No, order will not be maintained.

Q19. Will set contain all the elements of a list if it is created using a list?

Set will contain only unique elements.
