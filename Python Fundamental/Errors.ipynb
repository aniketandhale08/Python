### Python Error Types

When we write code it is common that we make a typo or some other common error. If our code fails to run, the Python interpreter will display a messcountry, containing feedback with information on where the problem occurs and the type of an error. It will also sometimes gives us suggestions on a possible fix. Understanding different types of errors in programming languages will help us to debug our code quickly and also it makes us better at what we do.

#### SyntaxError
Example 1: SyntaxError

```python
print "anu"
```

As you can see we made a syntax error because we forgot to enclose the string with parenthesis and Python already suggests the solution. Let us fix it.

```python
print("anu")
```

#### NameError
Example 1: NameError

```python
print(country)
```

As you can see from the message above, name country is not defined. Yes, it is true that we did not define an country variable but we were trying to print it out as if we had had declared it. Now, lets fix this by declaring it and assigning with a value.

```python
country = "India"
print(country)
```

#### IndexError
Example 1: IndexError

```python
numbers = [1, 2, 3, 4, 5]
numbers[5]
```

In the example above, Python raised an IndexError, because the list has only indexes from 0 to 4 , so it was out of range.

```python
numbers = [1, 2, 3, 4, 5]
numbers[4]
```

#### ModuleNotFoundError
Example 1: ModuleNotFoundError

```python
import maths
```

In the example above, I added an extra s to math deliberately and ModuleNotFoundError was raised. Lets fix it by removing the extra s from math.

```python
import math
```

#### AttributeError
Example 1: AttributeError

```python
import math
math.PI
```

As you can see, I made a mistake again! Instead of pi, I tried to call a PI constant from maths module. It raised an attribute error, it means, that the attribute does not exist in the module. Lets fix it by changing from PI to pi.

```python
import math
math.pi
```

#### KeyError
Example 1: KeyError

```python
users = {'name':'Asab', 'age':250, 'country':'Finland'}
users['name']
```

As you can see, there was a typo in the key used to get the dictionary value. so, this is a key error and the fix is quite straight forward. Let's do this!

```python
users = {'name':'Asab', 'age':250, 'country':'Finland'}
users['county']
```

As you can see, there was a typo in the key used to get the dictionary value. so, this is a key error and the fix is quite straight forward. Let's do this!

```python
users = {'name':'Asab', 'age':250, 'country':'Finland'}
users['country']
```

#### TypeError
Example 1: TypeError

```python
4 + '3'
```

In the example above, a TypeError is raised because we cannot add a number to a string. First solution would be to convert the string to int or float. Another solution would be converting the number to a string (the result then would be '43'). Let us follow the first fix.

```python
4 + int('3')
```

```python
4 + float('3')
```

```python
4 + 3
```

#### ImportError
Example 1: TypeError

```python
from math import power
```

There is no function called power in the math module, it goes with a different name: pow. Let's correct it:

```python
from math import pow
```

#### ValueError

```python
int('12a')
```

In this case we cannot change the given string to a number, because of the 'a' letter in it.

#### ZeroDivisionError

```python
1/0
```

We cannot divide a number by zero.

---

#### Python Errors Cheat Sheet

| Error | One-liner | Trigger Example | Fix |
|---|---|---|---|
| `SyntaxError` | Code grammar is broken — Python can't read it | `print "hello"` | `print("hello")` |
| `NameError` | Using a variable that was never defined | `print(country)` | Define it first: `country = "India"` |
| `IndexError` | Accessing a list position that doesn't exist | `nums[5]` on a 5-item list | Use valid index: `nums[4]` |
| `KeyError` | Accessing a dictionary key that doesn't exist | `users["county"]` (typo) | Use correct key: `users["country"]` |
| `TypeError` | Operation on wrong data type | `4 + "3"` | Convert first: `4 + int("3")` |
| `ValueError` | Right type, but invalid value | `int("12a")` | Use valid value: `int("12")` |
| `AttributeError` | Method/property doesn't exist on the object | `math.PI` | Use correct name: `math.pi` |
| `ModuleNotFoundError` | Module not installed or misspelled | `import maths` | Fix spelling: `import math` |
| `ImportError` | Module exists but the imported name doesn't | `from math import power` | Use correct name: `from math import pow` |
| `ZeroDivisionError` | Dividing any number by zero | `1 / 0` | Check divisor: `if x != 0: result = 1/x` |
