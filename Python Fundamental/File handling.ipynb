### File Handling

- File handling in Python is the process of creating, reading, writing, and updating files using code.
- Files can be in different formats like: .txt, .json, .xml, .csv, .tsv, .excel

- File handling is an import part of programming which allows us to create, read, update and delete files. In Python to handle data we use open() built-in function.

    ```python
    # Syntax
    open('filename', mode) # mode(r, a, w, x, t,b)  could be to read, write, update
    ```

    **File Modes (Important)**

    - "r" - Read - Default value. Opens a file for reading, it returns an error if the file does not exist  
    - "w" - Write - Opens a file for writing, creates the file if it does not exist  
    - "a" - Append - Opens a file for appending, creates the file if it does not exist  
    - "x" - Create - Creates the specified file, returns an error if the file exists  
    - "t" - Text - Default value. Text mode  
    - "b" - Binary - Binary mode (e.g. images)

### File Types

##### 📌 Dictionary vs JSON — Key Differences

- Where it lives → Dictionary lives in .py file, JSON lives in .json file  

- Quotes on keys → Dictionary accepts single or double quotes, JSON only double quotes  

- Type in Python → Dictionary is <class 'dict'>, JSON is <class 'str'>  

- Boolean values → Dictionary uses True/False, JSON uses true/false  

- Null values → Dictionary uses None, JSON uses null  

- Variable → Dictionary always needs a variable in Python, pure JSON has no variable  

- Outer quotes → Dictionary has no outer quotes, JSON string in Python can be written with outer quotes  

- Key types → Dictionary keys can be int, string, tuple etc, JSON keys can only be strings  

- Usage → Dictionary is used to work with data in code, JSON is used to transfer/store data  

##### File with json Extension

JSON stands for JavaScript Object Notation. Actually, it is a stringified JavaScript object or Python dictionary.

NOTE:
- Dictionary → accepts both single ' and double " quotes
- JSON → only double " quotes, strictly

````python
# Dictionary

person = {
    "name": "Alice",
    "age": 30,
    "skills": ["Python", "SQL"]
}

# Nested dictionary:
person = {
    "name": "Alice",
    "address": {
        "city": "Mumbai",
        "pin": 400001
    }
}

# JSON
{
    "name": "Alice",
    "address": {
        "city": "Mumbai",
        "pin": 400001
    }
}
````

- json.load() → used to read JSON from a file

1. Write a function that counts the number of lines and number of words in a text file.

   i. Read `obama_speech.txt` file and count the number of lines and words  
   ii. Read `michelle_obama_speech.txt` file and count the number of lines and words  
   iii. Read `donald_speech.txt` file and count the number of lines and words  
   iv. Read `melina_trump_speech.txt` file and count the number of lines and words  

````python
import os

def counting(file_paths):
    results = []

    for path in file_paths:
        line_count = 0
        words_count = 0

        try:
            with open(path, "r") as file:
                for line in file:
                    line_count += 1
                    words_count += len(line.split())

            file_name = os.path.basename(path)
            results.append((file_name, line_count, words_count))

        except FileNotFoundError:
            print(f"{path} not found")

    return results


files = [
    r"data\File handling\obama_speech.txt",
    r"data\File handling\michelle_obama_speech.txt",
    r"data\File handling\donald_speech.txt",
    r"data\File handling\melina_trump_speech.txt"
]

output = counting(files)

for file_name, lines, words in output:
    print(f"The {file_name} has {lines} lines and {words} words.")
````

````python
import json

def most_spoken_languages(filename, top_n):
    with open(filename, encoding="utf-8") as file:
        data = json.load(file)

    lang_count = {}
    

    for country in data:
        for lang in country["languages"]:
            if lang in lang_count:
                lang_count[lang] += 1
            else:
                lang_count[lang] = 1

    # sort by count (descending)
    sorted_lang = sorted(lang_count.items())
    sorted_lang.sort(key=lambda x: x[1], reverse=True)
    top_lang = sorted_lang[:top_n]

    result = []
    for lang, count in top_lang:
        result.append((count, lang))

    return result

print(most_spoken_languages(r"data\File handling\countries_data.json", 10))
````

5. Find the most common words in the English language. Call the name of your function find_most_common_words, it will take two parameters - a string or a file and a positive integer, indicating the number of words. Your function will return an array of tuples in descending order. Check the output.

````python
def find_most_common_words(filename, top_N):
    word_count = {}

    with open(filename, encoding="utf-8") as file:
        for line in file:
            words = line.split()

            for word in words:
                word = word.lower()

                if word in word_count:
                    word_count[word] += 1
                else:
                    word_count[word] = 1

    sorted_word = sorted(word_count.items(), key=lambda x: x[1], reverse=True)
    top_word = sorted_word[:top_N]

    result = []
    for word, count in top_word:
        result.append((count, word))

    return result

find_most_common_words(r"data\File handling\michelle_obama_speech.txt", 10)
````

````python
import os

def find_most_frequent_words(file_paths):
    final_result = {}

    for path in file_paths:
        word_count = {}

        with open(path, encoding="utf-8") as file:
            for line in file:
                words = line.split()

                for word in words:
                    word = word.lower()

                    if word in word_count:
                        word_count[word] += 1
                    else:
                        word_count[word] = 1

        sorted_word = sorted(word_count.items(), key=lambda x: x[1], reverse=True)
        top_word = sorted_word[:10]

        file_name = os.path.basename(path)

        result = []
        for word, count in top_word:
            result.append((count, word))

        final_result[file_name] = result

    return final_result

files = [
    r"data\File handling\obama_speech.txt",
    r"data\File handling\michelle_obama_speech.txt",
    r"data\File handling\donald_speech.txt",
    r"data\File handling\melina_trump_speech.txt"
]

output = find_most_frequent_words(files)

for file, data in output.items():
    print(f"{file} -> {data}")
````

7. Write a python application that checks similarity between two texts. It takes a file or a string as a parameter and it will evaluate the similarity of the two texts. For instance check the similarity between the transcripts of Michelle's and Melina's speech. You may need a couple of functions, function to clean the text(clean_text), function to remove support words(remove_support_words) and finally to check the similarity(check_text_similarity). List of stop words are in the data directory

````python
pip install nltk
````

````python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

def clean_text(text):
    text = text.lower()
    for ch in ".,!?;:\"()[]":
        text = text.replace(ch, "")
    return text

def check_text_similarity(file1, file2):
    stop_words = set(stopwords.words("english"))

    with open(file1, encoding="utf-8") as f1:
        text1 = f1.read()
    with open(file2, encoding="utf-8") as f2:
        text2 = f2.read()

    # clean text
    text1 = clean_text(text1)
    text2 = clean_text(text2)

    # split words
    words1 = text1.split()
    words2 = text2.split()

    # remove stop words
    words1 = [w for w in words1 if w not in stop_words]
    words2 = [w for w in words2 if w not in stop_words]

    # convert to sets
    set1 = set(words1)
    set2 = set(words2)

    # similarity
    common = set1.intersection(set2)
    similarity = len(common) / len(set1.union(set2)) * 100

    return f"Similarity: {similarity:.2f}"

print(check_text_similarity(
    r"data\File handling\michelle_obama_speech.txt",
    r"data\File handling\melina_trump_speech.txt"
))
````

9. Read the `hacker news CSV` file and find:

   i. Count the number of lines containing "python" or "Python"  
   ii. Count the number of lines containing "JavaScript", "javascript", or "Javascript"  
   iii. Count the number of lines containing "Java" and not "JavaScript"  

````python
import csv

def count_hacker_news(file):
    python_count = 0
    js_count = 0
    java_count = 0

    with open(file, encoding="utf-8") as f:
        reader = csv.reader(f)

        for row in reader:
            line = " ".join(row).lower()

            # i. python
            if "python" in line:
                python_count += 1

            # ii. javascript
            if "javascript" in line:
                js_count += 1

            # iii. java but not javascript
            if "java" in line and "javascript" not in line:
                java_count += 1

    return f"Python: {python_count} \nJavascript: {js_count} \nJava: {java_count}"

print(count_hacker_news(r"data\File handling\hacker_news.csv"))
````

### Python File Handling — Cheat Sheet

#### open() Modes
| Mode | Does |
|------|------|
| `r` | Read (default) — error if file missing |
| `w` | Write — overwrites, creates if missing |
| `a` | Append — adds to end, creates if missing |
| `x` | Create — error if already exists |

Always use `with open()` → auto-closes file.

#### Format Comparison
| | .txt | .json | .csv | .xlsx | .xml |
|--|------|-------|------|-------|------|
| Type | Plain text | String (in Python) | Plain text | Binary | Plain text |
| Structure | Unstructured | Key-value / nested | Tabular | Tabular + sheets | Hierarchical |
| Import | built-in | `json` | `csv` | `openpyxl` | `xml.etree.ElementTree` |
| Use case | Logs, notes | APIs, config | Data pipelines | Reports | APIs, config, data exchange |
