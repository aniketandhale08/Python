#### Reading from URL
By now you are familiar with how to read or write on a file located on you local machine. Sometimes, we would like to read from a website using url or from an API. API stands for Application Program Interface. It is a means to exchange structured data between servers primarily as json data. To open a network connection, we need a package called requests - it allows to open a network connection and to implement CRUD(create, read, update and delete) operations. In this section, we will cover only reading ore getting part of a CRUD.

````python
pip install requests
````

We will see get, status_code, headers, text and json methods in requests module:

- get(): to open a network and fetch data from url - it returns a response object
- status_code: After we fetched data, we can check the status of the operation (success, error, etc)
- headers: To check the header types
- text: to extract the text from the fetched response object
- json: to extract json data Let's read a txt file from this website

````python
import requests # importing the request module

url = 'https://www.w3.org/' # text from a website

response = requests.get(url) # opening a network and fetching a data

print(response)
print(response.status_code) # status code, success:200
print(response.headers)     # headers information
print(response.text) # gives all the text from the page
````

Let us read from an API. API stands for Application Program Interface. It is a means to exchange structure data between servers primarily a json data. An example of an API:https://restcountries.com/v3.1/name/peru. Let us read this API using requests module.

````python
import requests
import json

url = 'https://restcountries.com/v3.1/name/india'
response = requests.get(url)

print(response)
print(response.status_code)

countries = response.json()

print(json.dumps(countries[:1], indent=4))
````

We use json() method from response object, if the we are fetching JSON data. For txt, html, xml and other file formats we can use text.

1. Read this url and find the 10 most frequent words. romeo_and_juliet = 'https://www.gutenberg.org/files/1112/1112-0.txt'

````python
import requests
import re
from collections import Counter

url = "https://www.gutenberg.org/files/1112/1112-0.txt"
text = requests.get(url).text.lower()

words = re.findall(r'\b[a-z]+\b', text)

stopwords = set([
    "the","and","to","of","a","in","that","is","it","for","on","with","as","was","but","be"
])

filtered = []

for w in words:
    if w not in stopwords:
        filtered.append(w)

freq = Counter(filtered)

print(freq.most_common(10))
````

2. Read the cats API and cats_api = 'https://api.thecatapi.com/v1/breeds' and find :
    1. the min, max, mean, median, standard deviation of cats' weight in metric units.
    2. the min, max, mean, median, standard deviation of cats' lifespan in years.
    3. Create a frequency table of country and breed of cats

````python
import requests
import statistics

url = "https://api.thecatapi.com/v1/breeds"
data = requests.get(url).json()

weights = []
lifespans = []
country_freq = {}

# loop through each breed
for cat in data:
    
    # ---- WEIGHT ----
    weight_str = cat.get("weight", {}).get("metric", "")
    if weight_str != "":
        parts = weight_str.split(" - ")
        
        if len(parts) == 2:
            w1 = float(parts[0])
            w2 = float(parts[1])
            avg_weight = (w1 + w2) / 2
            weights.append(avg_weight)
    
    # ---- LIFESPAN ----
    life_str = cat.get("life_span", "")
    if life_str != "":
        parts = life_str.split(" - ")
        
        if len(parts) == 2:
            l1 = int(parts[0])
            l2 = int(parts[1])
            avg_life = (l1 + l2) / 2
            lifespans.append(avg_life)
    
    # ---- COUNTRY FREQUENCY ----
    country = cat.get("origin", "Unknown")
    
    if country in country_freq:
        country_freq[country] += 1
    else:
        country_freq[country] = 1

# ---- WEIGHT STATS ----
print("Weight Stats:")
print(min(weights), max(weights), statistics.mean(weights),
      statistics.median(weights), statistics.stdev(weights))

# ---- LIFESPAN STATS ----
print("\nLifespan Stats:")
print(min(lifespans), max(lifespans), statistics.mean(lifespans),
      statistics.median(lifespans), statistics.stdev(lifespans))

# ---- COUNTRY FREQUENCY ----
print("\nCountry Frequency:")
for k in country_freq:
    print(k, ":", country_freq[k])
````

3. Read the countries API and find: https://restcountries.com/v3.1/all?fields=name,languages
    1. the 10 most spoken languages
    2. the total number of languages in the countries API

````python
import requests

url = "https://restcountries.com/v3.1/all?fields=name,languages"
data = requests.get(url).json()

language_count = {}
unique_languages = set()

# loop through countries
for c in data:
    languages = c.get("languages")
    
    if languages is not None:
        for lang in languages.values():
            
            # count frequency
            if lang in language_count:
                language_count[lang] += 1
            else:
                language_count[lang] = 1
            
            # add to unique set
            unique_languages.add(lang)

# ---- sort languages by frequency ----
sorted_lang = sorted(language_count.items(), key=lambda x: x[1], reverse=True)

print("Top 10 Most Spoken Languages:")
for i in range(10):
    print(sorted_lang[i])

# ---- total languages ----
print("\nTotal Unique Languages:", len(unique_languages))
````

4. UCI is one of the most common places to get data sets for data science and machine learning. Read the content of UCL (https://archive.ics.uci.edu/ml/datasets.php). Without additional libraries it will be difficult, so you may try it with BeautifulSoup4

````python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

driver = webdriver.Chrome()
driver.get("https://archive.ics.uci.edu/datasets")

time.sleep(5)

titles = driver.find_elements(By.TAG_NAME, "h2")

for t in titles:
    print("Dataset:", t.text)
    
    parent = t.find_element(By.XPATH, "..")
    print(parent.text)
    print("-" * 50)

driver.quit()
````
